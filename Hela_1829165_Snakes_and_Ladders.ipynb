```python
import random          #importing for the random number generation

snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}
ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}

def get_name():                      #getting the names of the players
    print("*"*50)
    p1=None
    while not p1:
        
        p1=str(input("Player 1:- "))
    p2=None
    while not p2:
        p2=str(input("Player 2:- "))
        
    return p1,p2

def  roll_dice():                     #for rolling the dice and returning the obtained value
    inp=input();
            
    if inp.lower() == 'roll':         #for the random no generation
            k=random.randint(1,6)
            print("You got a "+str(k))
            return int(k)
    elif inp.isnumeric():
        if int(inp) >=1 and int(inp) <=20: #for the manual mode
            print("You got a "+str(inp))
            return int(inp)
        
            
        

        
        
def if_ladder_there(pname, currval):        #for returning the current position after the ladder traversal to snl fn
    print( pname+ "  ladder's up to" + str(currval))


        
def if_snake_there(pname,currval):        #for returning the current position after hitting a snake to snl fn
    print( pname+ "  snake down to " + str(currval))
    
def snake_and_ladder(dval,currval,pname):   #for returning the final position to the start function
    oldval=currval
    currval=currval+dval
    
    if currval>100:                         #for handling beyond  100 cases
        print("Can't go beyond 100 somry")
        return oldval
    if currval in ladders:                       #if the current value is in the ladders dictionary
        final_position = ladders.get(currval)
        if_ladder_there(pname,currval)
    elif currval in snakes:                      #if the current value is in the snakes dictionary
        final_position = snakes.get(currval)
        if_snake_there(pname,currval)
    
        
    else:
        final_position = currval
    
    return final_position

    
    
    
def start():
    print("#########Welcome to the game#########")
    
    pname1, pname2 = get_name()                    #for getting the names
    

    curr1 = 0                                      #setting the current values to 0
    curr2= 0

    while True:
        
        print(" " + pname1 + ": ")
        
        
        k= roll_dice()
            
        curr1 = snake_and_ladder(k,curr1,pname1)
        print(pname1,"is at "+str(curr1))

        if curr1 == 100:
            print(" ",pname1," Has won the game")
            print("#####Game Over#####")
            break

        print(" " + pname2 + ": ")
        
        
        s= roll_dice()
        curr2 = snake_and_ladder( s,curr2,pname2)
        print(pname2,"is at "+str(curr2))
        
 
        if curr2 == 100:
            print(" ",pname2," Has won the game")
            print("#####Game Over#####")
            break
 
    
    
if __name__ == "__main__":
    start()    
```
